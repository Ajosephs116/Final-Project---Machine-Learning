```python
import glob
import pandas as pd
import numpy as np
import pandas_profiling as pp
from functools import reduce
from sqlalchemy import create_engine
```

```python
data_frames = [pd.read_csv(f) for f in glob.glob('UTS_CSV_FILES/TMS_GS_*.csv')]

column_names = ['name', 'country_name', 'country_id', 'opponent_name', 'opponent_country_name', 'opponent_country_id', 
                'date', 'tournament', 'level', 'bestOf', 'surface', 'indoor', 'speed', 'round', 'score', 'winner']

pbjsandwich = reduce(
    lambda peanutbutter, jelly: pd.merge(
        peanutbutter, jelly, how='outer', on=column_names, validate="one_to_one"
    ), data_frames
)

pbjsandwich.head()
```

```python
print(len(data_frames))
```

```python
list(pbjsandwich.columns)
```

```python
pbjsandwich.dtypes
```

```python
objs_to_change = [
    'aces_percent', 'double_fault_percent', 'dbl_flt_per_2nd_Serv', '1st_Serve', '1st_serve_won', '2nd_serv_won', 
    'service_points_won', 'svc_inplay_pnts_won', 'break_points_saved', 'service_games_won', '1st_serve_return_won', 
    '2nd_serve_return_won', 'return_points_won', 'return_pnts_inplay_won', 'break_points_won', 'return_games_won', 
    'total_points_won_percent', 'total_2nd_serve_inplay_points_won', 'games_won_percent'
]
```

```python
pbjsandwich_obj = pbjsandwich.select_dtypes(['object'])
pbjsandwich_obj.head()
```

```python
pbjsandwich[pbjsandwich_obj.columns] = pbjsandwich_obj.apply(lambda x: x.str.strip('%'))
```

```python
pbjsandwich.head()
```

```python
pbjsandwich[objs_to_change] = pbjsandwich[objs_to_change].apply(pd.to_numeric)
```

```python
pbjsandwich.dtypes
```

```python
pbjsandwich.head()
```

```python
pbjsandwich.to_csv(path_or_buf="Resources/UTS_All.csv", index=False)
```

```python
pp.ProfileReport(pbjsandwich)
```

```python
engine = create_engine('sqlite:///Resources/Tennis_Collection.db', echo=False)
```

```python
pbjsandwich.to_sql('UTS_All', con=engine, index=False, if_exists='replace')
```
