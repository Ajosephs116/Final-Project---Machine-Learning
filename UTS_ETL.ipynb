```python
import glob
import pandas as pd
import numpy as np
from functools import reduce
```

```python
data_frames = [pd.read_csv(f) for f in glob.glob('UTS_CSV_FILES/TMS_GS_*.csv')]

column_names = ['name', 'country_name', 'country_id', 'opponent_name', 'opponent_country_name', 'opponent_country_id', 
                'date', 'tournament', 'level', 'bestOf', 'surface', 'indoor', 'speed', 'round', 'score', 'winner']

pbjsandwich = reduce(
    lambda peanutbutter, jelly: pd.merge(
        peanutbutter, jelly, how='outer', on=column_names, validate="one_to_one"
    ), data_frames
)

pbjsandwich.head()
```
